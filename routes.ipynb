```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df5 = pd.read_csv("route.csv",sep=';',decimal=',')
df5.head()
```

```python
df5.dtypes
```

```python
df5['epaisseurdernierstravaux'].dtypes
```

```python
df6=df5.dropna(how = 'any')
df6.head()
#nettoyage
```

```python
a=df6['naturedernierstravaux'].value_counts()
a.head()
#les 5 plus fréquents travaux
```

```python
df=df6.loc[:,['epaisseurdernierstravaux','anneedernierstravaux']]
df.head()
```

```python
dfa=df.groupby('anneedernierstravaux').mean()
dfa
```

```python
df1=df6.loc[:,['naturedernierstravaux','anneedernierstravaux']]
df1.head()
```

```python
dfb=df1.groupby('anneedernierstravaux').count()
dfb
```

```python

df3=pd.concat([dfa, dfb], axis = 2)
dfd=df3.sort_values(by = 'epaisseurdernierstravaux', ascending = False)
dfd
#le nombre de travaux réalisés et l’épaisseur moyenne
```

```python
%matplotlib inline
ax = dfb.plot(kind='bar', title ="Repartition",legend=True,color = ['blue'], fontsize=8)
ax
```

```python
%matplotlib inline
plt.pie(dfd['naturedernierstravaux'])
#representation en pie
```
